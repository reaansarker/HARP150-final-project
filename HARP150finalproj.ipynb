```python
#gameboard setup

the_chart = {'1': ' ' , '2': ' ' , '3': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '7': ' ' , '8': ' ' , '9': ' ' }

# function to print the updated board after every move

def printchart(chart):
    print(chart['1'] + '|' + chart['2'] + '|' + chart['3'])
    print('-+-+-')
    print(chart['4'] + '|' + chart['5'] + '|' + chart['6'])
    print('-+-+-')
    print(chart['7'] + '|' + chart['8'] + '|' + chart['9'])


chart_keys = []

for key in the_chart:
    chart_keys.append(key)


#game function
def game():

    player = 'X'
    count = 0
        

    for i in range(10):
        printchart(the_chart)
        print("Your turn, " + player + ". Where will you move?")

        #user entry
        move = input()        

        if the_chart[move] == ' ':
            the_chart[move] = player
            count += 1
        else:
            print("This space is already taken.\nWhere will you move?")
            continue

        #check if player X or O has won,for every move after 5 moves. 
        if count >= 5:
            if the_chart['1'] == the_chart['2'] == the_chart['3'] != ' ': # across the top
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")                
                break
            elif the_chart['4'] == the_chart['5'] == the_chart['6'] != ' ': # across the middle
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break
            elif the_chart['7'] == the_chart['8'] == the_chart['9'] != ' ': # across the bottom
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break
            elif the_chart['1'] == the_chart['4'] == the_chart['7'] != ' ': # down the left side
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break
            elif the_chart['2'] == the_chart['5'] == the_chart['8'] != ' ': # down the middle
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break
            elif the_chart['3'] == the_chart['6'] == the_chart['9'] != ' ': # down the right side
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break 
            elif the_chart['7'] == the_chart['5'] == the_chart['3'] != ' ': # diagonal
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break
            elif the_chart['1'] == the_chart['5'] == the_chart['9'] != ' ': # diagonal
                printchart(the_chart)
                print("\nGame Over!\n")                
                print(" Congrats " +player + ", you won!")
                break 
                
                #change the player after every move.
        if player =='X':
            player = 'O'
        else:
            player = 'X' 

        #result is a tie if neither win.
        if count == 9:
            print("\nGame Over!\n")                
            print("No winner- It's a tie!")
             
      
    
    #restart game or not.
    restart = input("Do want to play Again?(yes/no)")
    if restart == "yes" or restart == "Yes":  
        for key in chart_keys:
            the_chart[key] = " "

        game()

if __name__ == "__main__":
    game()
```
